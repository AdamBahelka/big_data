### 1)Random Numbers

Generate 5 **lists** of length 1M of random numbers between 0 and 10 (Think of it as the number of tweets for 1M customers for everyday through 5 days)

1. Calculate the number of tweets per customer per 5 days.
2. Calculate the total number of tweets per user and plot its histogram. What distribution do you observe? Explain why? 
3. What is the index of the customer with the maximum number of tweets?
4. Time and report each step of your code.



```python
#Generating the number of tweets
day1 = []
day2 = []
day3 = []
day4 = []
day5 = []
t1 = time.time()
for i in range(1000000):
  day1.append(random.randint(0,10))
  day2.append(random.randint(0,10))
  day3.append(random.randint(0,10))
  day4.append(random.randint(0,10))
  day5.append(random.randint(0,10))
t2 = time.time()
print('Using lists, generating the number of tweets for 1M users during the 5 days takes {t:f} seconds.'.format(t=t2-t1))
```

```python
total_tweets_per_user = []
t1 = time.time()
for i in range(1000000):
    total_tweets_per_user.append(day1[i]+day2[i]+day3[i]+day4[i]+day5[i])
t2 = time.time()
print('Using lists, finding the total number of tweets per user takes {t:f} seconds.'.format(t=t2-t1))
```

```python
plt.hist(total_tweets_per_user,bins=100);
```

### 2) Pokemon Dataset

Download the [Pokemon images](https://www.kaggle.com/datasets/thedagger/pokemon-generation-one?resource=download) to your computer. (you can use google account to log into Kaggle.) and write a Python code to perform the following:
1. Create a dictionary that contains for every pokemon a dictionary with the following information: name of the pokemon, dimensions of the image, and total number of pixels. Every pokemon image should be added to the dictionary under its file name. How long did it take to create the dictionary? 
2. Using the json module, save the pokemon dictionary as JSON file.

During the process if you face any problems, avoid them but also record those problems. To read images use the Pillow (PIL) library.

```python
pokemon_dict = {}
unsupported_images = []
t1 = time.time()
for pokemon_folder in os.listdir('/home/atatar/Documents/CIT/Teaching/UCG/DataScience/week1/Data/pokemon/dataset/'):
    folder = os.path.join('/home/atatar/Documents/CIT/Teaching/UCG/DataScience/week1/Data/pokemon/dataset/',pokemon_folder)
    for pokemon_image in os.listdir(folder):
        try:
            pokemon = {}
            pokemon_id = pokemon_image.split('.')[0]
            file_path = os.path.join(folder,pokemon_image)
            img = Image.open(file_path)
            w,h = img.size
            total_pix = w*h
            pokemon['pokemon name'] = pokemon_folder
            pokemon['image dimensions'] = (w,h)
            pokemon['pixels'] = total_pix
            pokemon_dict.update({pokemon_id:pokemon})
        except UnidentifiedImageError:
            unsupported_images.append(pokemon_image)

t2 = time.time()

```

```python
print('Time to create the pokemon dictionary is {t:f} seconds.'.format(t=t2-t1))
```

```python
print('Number of problems is {n}.'.format(n=len(unsupported_images)))
```

```python
with open('pokemon.json','w') as fout:
    json.dump(pokemon_dict,fout)
```
