## Linux Commands
We can use linux command in Jupyter notebook by preceding the command with the ! sign.

### Looking into Directories
We look into the directories with the `ls` command.

```python
# printing the working directory
! pwd
```

```python
#content of the working directory
! ls
```

We can use flags to modify the results of the commands:

* -l (l for long) lists the details of the files
* -a (a for all) list all the files (even the hidden ones)
 
We can also combine flags

```python
#detailed content of the directory
! ls -l
```

```python
! ls -al
```

We can chage the directory with the `cd`.

```python
# going one level up
! cd ..
```

```python
# going to a new directory
! cd /home/atatar/Documents/Marketing
```

```python
# going to the directories under your current working directory
! cd ./EmptyFolder
```

```python
# listing the content of another folder
! ls ./EmptyFolder
```

```python
# listing the content of another folder
! ls /home/atatar/Documents/Marketing/SLIM
```

### Manipulating Files and Directories

```python
# show the content of the file
! cat ./LinuxPractice/mycsvfiles.txt
```

```python
# write the content of the file into a new file.
# if the file does not exist it will be created.
# if the file exists, its conetnt will be replaced.
! cat ./LinuxPractice/mycsvfiles.txt > allfiles.txt
```

```python
# if the file exists, append to the end of it.
# if the file does not exist, craete a new file.
! cat ./LinuxPractice/myexcelfiles.txt >> allfiles.txt
```

```python
# copy a file
! cp allfiles.txt ./LinuxPractice
```

```python
# create a directory
! mkdir ./Folder1
```

```python
! ls ./LinuxPractice
```

```python
# copy all the files in a directory
! cp -r ./LinuxPractice/ ./Folder1
```

```python
# copy all the files in a directory
! cp -r ./LinuxPractice/* ./Folder1
```

```python
# move files
! mkdir ./Folder2
! mv allfiles.txt ./Folder2
```

```python
! mv ./Folder1/mycsvfiles.txt ./Folder2/mycsvfiles2.txt
```

```python
#changing name of a file = moving a file
! mv ./LinuxPractice/allfiles.txt ./LinuxPractice/myallfiles.txt
```

```python
# move a directory
! mv -r ./Folder1 ./Folder2
```

```python
# move a directory
! mv ./Folder1/* ./Folder2
```

```python
# delete a file
# BECAREFUL YOU CANNOT UNDO THIS OPERATION!!!!!!!!!!!!!!!
! rm ./LinuxPractice/myallfiles.txt
```

```python
#delete a directory
! rm -r Folder1
```

```python
# delete a directory
! rm -r Folder2
```
