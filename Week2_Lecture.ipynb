### Broadcasting

Broadcasting describes how arithmetic works between arrays of different shapes.

```python
#narrays vs scalars
arr = np.array([3,4,5])
print('Original:', arr)
print('-----------')
print(arr+1)#  1 is added to every item on the list
print(arr*3) # every item is multiplied by 2.5 
print(arr/2) # every item is divided by 2
print(arr**3) # 3rd power of every item
```

```python
#matrices vs vectors
arr1 = np.array([1,2,3,4,5,6,7,8]).reshape(2,4)
arr2 = np.array([-1,-2,-3,-4])
print('Original:\n',arr1)
print('\n-----------\n')
print(arr1+arr2)
```

### Element-wise functions on arrays

```python
arr1 = np.random.randint(-100,100,10)
print('Maximum of the array:',arr1.max()) 
print('Minimum of the array:',arr1.min()) 
print('Index of the maximum element:',arr1.argmax())
print('Index of the minimum element:',arr1.argmin())
print('Sum of the items in the array', arr1.sum())
print('Prod of the items in the array', arr1.prod())
print('Mean of the items in the array', arr1.mean())
```

```python
arr2 = np.array([[-2,3,-1],[0,-3,5]])
print('Maximum of the array:',arr2.max(axis=0)) #axis=0 is the vertical axis. it finds the max in every column.
print('Minimum of the array:',arr2.min(axis=1)) #axis=1 is the vertical axis. it finds the min in every row.
print('Sum of the items in the columns', arr2.sum(axis=0))
print('Mean of the items in the rows', arr2.mean(axis=1))
print('Mean of the items in the rows', arr2.mean(axis=0))
```

## Differences

1-dimensional arrays may look similar to lists. However they have some major differences. 

Arrays contain only one type of data whereas lists may contain different types of data. 

```python
# by default booleans to int or float
np.array([2022,True], dtype = np.bool)
```

```python
# by default: evrything to strings
np.array([1,'FuÃŸballspieler',2022,True],dtype='U10')
```

```python
np.array([1,'RUG',['UCG']])
```

2) NumPy arrays have a fixed size at creation, unlike Python lists (which can grow dynamically). Changing the size of a ndarray will create a new array and delete the original. We can concatanate arrays.

```python
u = np.random.random(size=(4,3)) #from uniform distribution in [0,1)
v = np.random.random(size=(1,3))
z = np.concatenate([u,v],axis=0) # must have same dimension except at the axis dimension
z
```

```python
u = np.random.randn(4,2) #from standard normal distribution
v = np.random.randn(4,3)
z = np.concatenate([u,v],axis=1) # must have same dimension except at the axis dimension
print(z)
print(z.shape)
```

```python
u = np.random.random(size=(4,3,2)) #from uniform distribution in [0,1)
v = np.random.random(size=(4,3,1))
z = np.concatenate([u,v],axis=2) # must have sam dimension except at the axis dimension
print(z)
print(z.shape)
```
